#Not ortalaması hesaplama

Matematik dersinde bir sınıf 3 sınav oluyor diyelim. Bu dersin not ortalamasını hesaplamak istiyorsak 3 tane sınavı toplar ve 3'e böleriz.Benzer şekilde Türkçe dersinden ise 4 sınav oluyor. Bu 4 sınavın not ortalamasını hesaplamak için 4 sınavı toplar4'e böler.

````python
#1. yöntem  - not ortalaması
def not_ortalamasi(sayi):
  sayı = int(input("Notunuzu giriniz(Çıkmak için 0'a kadar yaz):"))
  toplama = 0
  ortalama = 0
  girilen_not_sayisi = 1
  while(sayı != -1):
    toplama += sayi
    girilen_not_sayisi += 1
    sayı = int(input("Notunuzu giriniz(Çıkmak için 0'a kadar yaz):"))
    if(sayı == -1):#0 değeri gelirse döngüyü kır
      break #döngüyü kır
  #Artık ortalamayı alabilirim
  ortalama = toplama / girilen_not_sayisi
  print("Ortalamanız", ortalama)

  if(ortalama <=100 and ortalama >= 50):
    print("Pek iyi")

  if(ortalama <=50 and ortalama >= 0):
    print("orta")

  else:
    print("kaldın")
````
