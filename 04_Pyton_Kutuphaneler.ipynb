#Kütüphaneler

Hazır önceden birçok özel işlemi içinde bulunduran kullanıma hazır yapılardır
Phyton'da sayısız kütüphane vardır.  Örneğin:

*   Matamatiksel işlemler için **math**
*   Rastgele içerik ve sayı üretmek için **random**
*   İşetim sistemi dosyalarına erişmek için **os** kötöphanesi
*   Oyun geliştirnek için **turtle** veya **pygame** kütüphaneleri
*   Arayüz programlaması için **tkiter** kütüphanesi kullanılır.

##Peki kütüphaneler nasıl eklenir?

Küt. eklemek için genelde kullanılan yöntem şöyle:
import kütüphane_adı

Örneğin;
import random

```python
#Küt. her zaman projenin en başına eklenir.
import random #Rastgele içerik veya sayı üretmek için kullanılır.

rastgele_sayı = random.randint(1,11) #1-11 arası rastgele sayı seç
print(rastgele_sayı)
```

```python
#Küt. her zaman projenin en başına eklenir.
import random #Rastgele içerik veya sayı üretmek için kullanılır.

rastgele_sayı = random.randint(1,11) #1-11 arası rastgele sayı seç
print(rastgele_sayı)
if rastgele_sayı == 1:
  print("Uğur")
elif rastgele_sayı == 2:
  print("Göktuğ")
elif rastgele_sayı == 3:
  print("Azad")
elif rastgele_sayı == 4:
  print("Ebubekir")
elif rastgele_sayı == 5:
  print("Yusuf")
elif rastgele_sayı == 6:
  print("Ali")
elif rastgele_sayı == 7:
  print("Kerem")
elif rastgele_sayı == 8:
  print("Alikerem")
elif rastgele_sayı == 9:
  print("Kaya")
elif rastgele_sayı == 10:
  print("Mehmet")
elif rastgele_sayı == 11:
  print("Muhammet")
```

```python
import random as rd
import math as m

sayı1 = (1)

print(sayı1)

#Fonksiyonlar

print(m.factorial(sayı1))
print(m.comb(sayı1, 2))
print(m.perm(sayı1, 2))
print(m.sqrt(sayı1))
print(m.floor(sayı1))
print(m.pow(sayı1, 4))
print(m.log(sayı1, 3))
print(m.pi)
print(m.e)
```
