#Listelere Input İle Eleman Ekleme ve Çıkarma
Input kullanarak listelere eleman ekleyip çıkarabiliriz. Bazı programlama dillerinde add (ekleme) veya append (listeye içine ekleme) anlamına gelen bir fonksiyon ile ekleme yapılabilir. <br>

Çıkarma işlemlerinde delete (silme) veya remove (listeden çıkarma) anlamında bu fonksiyonlar kullanılabilir. İkisinden örnek vermek gerekirse liste.append(sayi) veya liste.remove(3) gibi farklı şekillerde kullanılabilir.

```python
tatlilar = ["traliçe","kemal paşa","sütlaç","pudink","baklava","pasta","kadayıf"]
tatli = input("Lütfen tatlı giriniz")
tatlilar.append(tatli)
print(tatlilar)
#Listeden ekleni silebilirim
tatlilar.remove(tatli)
print(tatlilar)
print("silme işlemi")
silinecek = input("Listeden silinecek tatlıyı seçin:")
tatlilar.remove(silinecek)
print("Ekstra silmeden sonra:")
print(tatlilar)
```

#Listelerde farklı işlemler
listeleri temizleme, elemen seçip aynı anda çıkarkma, listeyi tersten yazdırma,liste lelman sayısı veya indisi bilgisi alma gibi işlemler yapabiliriz.

```python
# Clear komutu -> Her şeyi siler.
haftanın_günleri = ['Pzt.','Sal.','Çar','Per.','Cum.','Cumartesi','Pazar']
gun = haftanın_günleri.count('Sal.')
print(gun)
```

```python
#Index komutu -> Yazılan Elemenı Sayar
haftanın_günleri =  ['Pzt.','Sal.','Çar','Per.','Cum.','Cumartesi','Pazar']
print(haftanın_günleri.index('Çar.'))
```

```python
#pop kumutu -> listenin içinden bahsedilen index'li elemanları yazar ve kisteden çıkartır.
haftanın_günleri =  ['Pzt.','Sal.','Çar','Per.','Cum.','Cumartesi','Pazar']
gun = haftanın_günleri.pop(3)
print(gun)
print(haftanin_gunleri)
```

```python
#Reverse komutu -> Listeyi tersine çevirir
haftanın_günleri =  ['Pzt.','Sal.','Çar','Per.','Cum.','Cumartesi','Pazar']
haftanin_gunleri.reverse()
print(haftanın_günleri)
```

#Listelerde Arama Yapma ve Döngü Kullanımı
Listeler içerisinde belirli koşullar sağlandığında yazdırılacak veya işe yarayacak bilgileri alabilmek için listelerde arama yapılmalıdır.
Listelerde arama yapmak için döngüler kullanılır.

```python
from os import EX_PROTOCOL
# 0 - 250 arasında 3'er 3'er sayıları yazdırma
ucerli_liste = []
onbesin_kati = []
for a in range(0,250,3):
  ucerli_liste.append(a)
print(ucerli_liste)
print("toplam eleman sayısı",len(ucerli_liste))
#bu listenin içinden 5'e bölünen sayılarıekrana yazdırma
for eleman in range(0,len(ucerli_liste)-1):
  if ucerli_liste[eleman] % 5 == 0:
    onbesin_kati.append(ucerli_liste[eleman])
print(onbesin_kati)
print("toplam eleman sayısı",len(onbesin_kati))
```
