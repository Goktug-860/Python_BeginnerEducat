#Değişkenlerin Özellikleri


1.  Değişkenler amacına göre isimlendirilir.
2.  İsimler, sayı, metin, kesirli sayı, doğru\yanlış değerlerini
alabilir.
3.  İsim verilirken Türkçe karakter olmamalı.

```python
#Farklı değişken türliri
araba_isim="TESLA" # metin sitrink demek.
arbamotor_güç=400 #sayı integer demek.
arba_yükseklik=123.9 #yükseklik float demek.
arba_genişlik=95.9 #Ondalık sayı Doble demek.
araba_çalışıyormu=True #doğru yanlış demek
araba_pahalı= False #doğru yanlış demek
print(araba_isim,arbamotor_güç,arba_yükseklik,arba_genişlik,araba_çalışıyormu,araba_pahalı)
```
