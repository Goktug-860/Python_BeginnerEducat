```python
sayı1 = int(input("1. sayıyı gir:"))
sayı2 = int(input("2. saıyyı gir:"))

toplam = sayı1 + sayı2
cıkartma = sayı1 - sayı2
çarpma = sayı1 * sayı2
bolme = sayı1 / sayı2
print(f"Toplam:{toplam},Çıkartma:{cıkartma},Çarpma:{çarpma},Bölme:{bolme}")
```
